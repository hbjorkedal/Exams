```python
import numpy as np
```

```python
λ = 50  # arrival rate 
μ1 = 30  # service rate at microphone 
μ2 = 40  # service rate at food preparation 

# Linear system of equations
A = np.array([
    [-λ, 0, 0, μ2, 0, 0, 0, 0],     # eq1
    [λ, -(λ + μ1), 0, 0, μ2, 0, 0, 0], # eq2
    [0, λ, -μ1, 0, 0, μ2, 0, 0],     # eq3
    [0, μ1, 0, -(λ + μ2), 0, 0, μ2, 0], # eq4
    [0, 0, μ1, λ, -(λ + μ1 + μ2), 0, 0, μ2], # eq5
    [0, 0, 0, 0, λ, -(μ1 + μ2), 0, 0], # eq6
    [0, 0, 0, 0, μ1, 0, -(λ + μ2), 0], # eq7
    [0, 0, 0, 0, 0, μ1, λ, -μ2],     # eq8
    [1, 1, 1, 1, 1, 1, 1, 1]         # Normalization
])

# Constants on the right-hand side of the equations
b = np.array([0, 0, 0, 0, 0, 0, 0, 0, 1])
```

The matrix is generated by Chat-GPT 4

```python
# Solve the system of linear equations
x = np.linalg.lstsq(A, b)
for i, eq in enumerate(x[0]):
    print(f"p{i+1} = {eq:.4f}")
```
